 ## with the Post Pruning method

```python
import pandas as pd
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
X=df.iloc[:,:-1]
```

```python
X
```

```python
from sklearn.datasets import load_iris
dataset= load_iris()
y = dataset.target
```

```python
y
```

```python
# do not use standadization or normalization 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
treeclassifire = DecisionTreeClassifier()
```

```python
treeclassifire.fit(X_train,y_train)
```

```python
## visulize the tree
import matplotlib.pyplot as plt
from sklearn import tree
plt.figure(figsize=(50,40))
tree.plot_tree(treeclassifire,filled=True)
```

```python
## post pruning with max_width 2 becuse the see the above tree 
treeclassifire= DecisionTreeClassifier(max_depth=2)
treeclassifire.fit(X_train,y_train)
plt.figure(figsize=(20,15))
tree.plot_tree(treeclassifire,filled=True)
```

```python
y_pred=treeclassifire.predict(X_test)
y_pred
```

```python
y_test
```

```python
y_pred==y_test
```

```python
from sklearn.metrics import accuracy_score,classification_report
```

```python

print(accuracy_score(y_pred,y_test))
print(classification_report(y_pred,y_test))
```
