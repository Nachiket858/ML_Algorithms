```python
from sklearn.datasets import make_blobs
X,y= make_blobs(n_samples=1000,n_features=3,centers=3, random_state=34)
```

```python
import matplotlib.pyplot as plt 
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=42)
```

```python
plt.scatter(X[:,0],X[:,1], c= y)
X
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for k in range(1,10):
    means = KMeans(n_clusters=k,init='k-means++')
    means.fit(X_train)
    wcss.append(means.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,10),wcss)
```

```python
kmeans = KMeans(n_clusters=3,init='k-means++')
y_pred  = kmeans.fit_predict(X_train)
```

```python
plt.scatter(X_train[:,0],X_train[:,1], c= y_pred)
```

```python
## calculating the value of k using the knee locatore
from kneed import KneeLocator
knee = KneeLocator(range(1,10),wcss,curve='convex',direction='decreasing')
knee.elbow
```

```python
## performance metrix
from sklearn.metrics import silhouette_score
silhouette_score(X_train,kmeans.labels_) 
```
